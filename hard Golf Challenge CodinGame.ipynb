```python
moves = {i: [np.array(a) for a in [[i,0], [0,-i], [0,i], [-i,0]]] for i in range(1, 9+1)}
moves
def paint(arr, fr, mv):
    arr = arr.copy()
    d = abs(mv[0]) + abs(mv[1])
    #print(fr, mv)
    i,j = fr
    mi,mj = mv
    #print(i,j,mi,mj)
    if i+mi >= arr.shape[0] or j+mj >= arr.shape[1] or i+mi < 0 or j+mj < 0:
        return None, None
    gridnoland = ((arr+1)//2 != 0) if d > 1 else (arr != 0)
    gridnofly = arr >= 0
    gridnolandorfly = gridnoland + gridnofly
    theend = arr[i+mi, j+mj] == 0
    if mi:
        part = gridnofly[i+np.sign(mi):i+mi:np.sign(mi), j+mj]
        if gridnoland[i+mi,j+mj] or part.any():
            return None,None
        arr[i+np.sign(mi):i+mi+np.sign(mi):np.sign(mi), j+mj] = d
    if mj:
        part = gridnofly[i+mi, j+np.sign(mj):j+mj:np.sign(mj)]
        if gridnoland[i+mi,j+mj] or part.any():
            return None,None
        arr[i+mi, j+np.sign(mj):j+np.sign(mj)+mj:np.sign(mj)] = d
    arr[i+mi, j+mj] = 100
    if theend:
        arr[i+mi, j+mj] = 200
        return arr, None
    return arr, [i+mi, j+mj]

def explore(n, ex=None, pos=None, tot=None):
    if not n:
        return tot
    if ex is None:
        max_dist = (n*(n+1))//2
        mid = [max_dist, max_dist]
        ex = np.ndarray([max_dist+1+max_dist, max_dist+1+max_dist], dtype=int) * 0-1
    else:
        ex = ex.copy()
    pos = pos if pos is not None else mid
    ex[pos[0],pos[1]] = 99
    if tot is None:
        tot = ex.copy()
    for m in moves[n]:
        newarr, newpos = paint(ex, pos, m)
        if newarr is not None:
            tot[newarr!=0] = np.maximum(newarr[newarr!=0], tot[newarr!=0])
            #tot[newarr!=0] += newarr[newarr!=0]
        if newpos:
            tot = explore(n-1, ex=newarr, pos=newpos, tot=tot)
    return tot

t = explore(3)
#assert np.max(np.abs(t-t.T)) == 0
t
```
